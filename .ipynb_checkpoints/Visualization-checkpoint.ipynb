```python
import geopandas as gpd
import geoplot as gplt
%matplotlib inline
```

```python
contiguous_usa = gpd.read_file(
  gplt.datasets.get_path('contiguous_usa')
)
contiguous_usa = contiguous_usa.sort_values('state').reset_index(drop=True) # Alphabetize the states list
contiguous_usa.head(3)
```

```python
import geoplot.crs as gcrs # (projections)
```

```python
gplt.polyplot(contiguous_usa, projection=gcrs.AlbersEqualArea())
```

```python
gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.longitude, df.latitude))
gdf
```

```python
ax = gplt.polyplot(
    contiguous_usa,
    projection=gcrs.AlbersEqualArea()
)
gplt.pointplot(gdf, ax=ax)
```

Look into those three anomalous points.

```python
mistaken_locations = gdf.query('latitude > 44 & longitude > -65')
mistaken_locations[['state_code', 'latitude', 'longitude', 'zip_code','city','Unnamed: 6']]
```

```python
gdf2 = gdf.query('latitude < 44 & longitude < -65') # Sift out the poor locations for now
```

```python
ax = gplt.polyplot(
    contiguous_usa,
    projection=gcrs.AlbersEqualArea(),
    figsize=(30,20)
)
gplt.pointplot(gdf2, ax=ax)
```

```python
status_bool = [1 if x == 'acquired' else 0 for x in gdf2.status]
gdf3 = gdf2.assign(success = status_bool)[['state_code','success']]
gdf3
```

```python
gdf3.groupby('state_code').agg('sum')
```

```python
success_rate_by_state = gdf3.groupby('state_code').agg('mean').reset_index()
success_rate_by_state
```
